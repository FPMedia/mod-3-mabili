```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Loading the dataset

df = pd.read_csv('Finance_data.csv')
```

```python
#calculating the mean of the ages and Fixed deposits of each individual in the dataset

mean1 = np.mean(df["age"])
mean2 = np.mean(df["Fixed_Deposits"])
```

```python
# calculating the median of the age and fixed deposits in the dataset

median1 = np.median(df["age"])
median2 = np.median(df["Fixed_Deposits"])
```

```python
# visualising the statistical information with the numpy operationS to show the mean values

# Define the labels and mean values
labels = ['age', 'Fixed_Deposits']
means = [mean1, mean2]

# Create a bar plot
plt.figure(figsize=(5, 4))
plt.bar(labels, means, color=['blue', 'green'], label = labels)
plt.xlabel('columns')
plt.ylabel('Mean Values')
plt.title('Mean Values of age and fixed deposits')
plt.legend()
plt.show()
```

```python
# standard deviation of age and fixed deposits
std1 = np.std(df["age"])
std2 = np.std(df["Fixed_Deposits"])
```

```python
#Visualising the standard deviation of age and fixed deposits

# Define the labels and median values

labels = ['age','Fixed_Deposits']
stds = [std1, std2]
    
    # Create a bar plot
plt.figure(figsize=(5, 4))
plt.bar(labels, stds, color=['blue', 'green'], label = labels)
plt.xlabel('Columns')
plt.ylabel('standard deviation Values')
plt.title('standard deviation Values of age and fixed deposits')
plt.legend()
    
plt.show()
```

```python
# calculating the correlation between the two columns

correlation_coefficient = np.corrcoef(df["age"], df["Fixed_Deposits"])[0, 1]
```

```python
# Create a scatter plot which will make it easier to see the correlation

plt.figure(figsize=(5, 4))
plt.scatter(df["age"], df["Fixed_Deposits"], color='blue', alpha=0.6, label = 'Data points')
plt.title('Correlation between Age and Fixed Deposits')
plt.xlabel('Age')
plt.ylabel('Fixed Deposits')

plt.legend()
plt.show()
```
